```python
import numpy as np
import matplotlib.pyplot as plt

# Generate earthquake magnitudes
a = 8  # constant related to total number of earthquakes
b = 1  # b-value, typically close to 1 for seismically active regions
M = np.linspace(0, 9, 1000)  # range of magnitudes
N = 10**(a - b*M)  # number of earthquakes with magnitude >= M

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.semilogy(M, N, 'b-', linewidth=2)
plt.xlabel('Magnitude (M)')
plt.ylabel('Number of earthquakes with magnitude  M')
plt.title('Gutenberg-Richter Law: Earthquake Magnitude Distribution')
plt.grid(True)
plt.show()
```
