```python
#### Example of a Histogram with Grocery price data:
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

groceries = np.array([
 24.72, 159.62, 154.89,   2.47,  76.04,  20.56,  87.66,  18.42,  34.72, 126.4,
 30.94, 170.11, 190.04,  77.04,   2.53,  97.24, 172.03,  22.86, 207.81,  32.14,
 130.98,  46.23,  56.17,  97.26,  27.95, 159.58,  52.82,  16.07,  63.71,  16.44,
 29.21, 162.77,  24.1,  142.74, 149.31,  23.27,  94.95, 187.04,  31.82,  31.69,
 203.31,  18.97,  24.37,  40.4,  177.31, 125.7,   61.3,   33.2,  138.3,  102.29,
 32.01,  23.4,  116.69,  23.24,  29.89,  29.88, 147.8,   21.27, 109.89, 151.52,
 23.69,  59.09,   8.67, 109.7,   30.71, 177.04,  86.57,  29.57, 159.04,  66.31,
 21.89, 120.39,  17.98, 134.57,  23.74,  32.24,  29.49,  85.49,  31.52,  75.93,
 25.27,  30.93,  27.77, 105.34,  36.58,  11.6,   37.51,  27.69, 166.9,   70.01,
  9.49,  35.61, 127.8,  122.91,  32.64,   7.6,   27.64, 143.6,   33.57, 146.21,
 141.25,  17.58, 144.67,  31.93, 124.73,  22.57,  27.79,  17.78, 157.52,  66.37,
 141.2,   86.88,  48.14,  32.05,   5.23,  77.61, 165.58,  41.77,  14.16, 137.91,
 178.62, 152.09
])

full_grocery_trips = groceries[groceries > 100]

# Robust method of getting right number of bins - used in a lot of statistics
def freedman_diaconis(data):
    iqr = stats.iqr(data)
    n = len(data)
    bin_width = 2 * iqr * (n ** (-1/3))
    return np.ceil((np.max(data) - np.min(data)) / bin_width)

groceries_to_plot = full_grocery_trips
print(f'full grocery trips: {len(full_grocery_trips)}')
print(f'per month: {len(full_grocery_trips) / 12}')

num_bins = int(freedman_diaconis(groceries_to_plot)) # wrap in int to remove decimal for num bins
mean_price = np.mean(groceries_to_plot)
median = np.median(groceries_to_plot)

# The degree of skewness can be quantified by calculating the difference between the mean and median. 
# A larger difference indicates a more skewed distribution, while a smaller difference suggests a more symmetric distribution.
# A negative skew means skew is to the left, positive number skew means the skew is to the right - 0 is symmetric (no skew)
skewness = stats.skew(groceries_to_plot)
plt.text(0.95, 0.8, f'Skewness: {skewness:.2f}', transform=plt.gca().transAxes,
         verticalalignment='top', horizontalalignment='right')

plt.hist(groceries_to_plot, bins=num_bins, color='lightgreen', edgecolor='black', label=f'Median: {np.round(median)}')
plt.axvline(mean_price, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {np.round(mean_price)}') # draw vertical line for mean
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Full Grocery Trips')
plt.legend()
plt.show()

groceries_to_plot = groceries
num_bins = int(freedman_diaconis(groceries_to_plot))
mean_price = np.mean(groceries_to_plot)
median = np.median(groceries_to_plot)

skewness = stats.skew(groceries_to_plot)
plt.text(0.95, 0.8, f'Skewness: {skewness:.2f}', transform=plt.gca().transAxes,
         verticalalignment='top', horizontalalignment='right')

plt.hist(groceries_to_plot, bins=num_bins, color='lightgreen', edgecolor='black', label=f'Median: {np.round(median)}')
plt.axvline(mean_price, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {np.round(mean_price)}') # vertical line for mean
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of All Grocery Trips')
plt.legend()
plt.show()
```

```python
## create data for the histogram

# number of data points
n = 1000

# generate data - log-normal distribution
data = np.exp( np.random.randn(n)/2 )
```

```python
# show as a histogram

# number of histogram bins
k = 5

plt.hist(data,bins=k)
plt.show()
```

```python
# another option
y,x = np.histogram(data,bins=k)

# bin centers
xx = (x[1:]+x[:-1])/2
print(type(xx))

plt.plot(xx,y)
plt.show()
```
