```python
########## DEGREES OF FREEDOM ################

import numpy as np
import matplotlib.pyplot as plt

def f(m,x,c):
    return m * x - c # mx + c

m = 5
c = 3
xx = np.linspace(-5,100,106)

SAMPLE_LEN = 6
first_6_x = xx[:SAMPLE_LEN]
print(f'first 6 x vals: {first_6_x}')
first_output = f(m,xx[0],c)
print(f'first output: {f(m,xx[0],c)}')

# One Degree of Freedom (x is free, m and c are constant)
plt.plot(xx,f(m,xx,c),color='green', label='One degree of freedom (x)')


# two degrees of freedom (m is free and x is free, c is fixed)
random_m = np.random.randint(11, 20, size=SAMPLE_LEN)
print(f'2 dof: {random_m=}')

for i in range(SAMPLE_LEN):
    plt.plot(xx,f(random_m[i],xx,c),color='magenta', label='Two degrees of freedom (m,x)' if i == 0 else "")


# three degrees of freedom (m is free, x is free, and c is free)
random_m = np.random.randint(21, 30, size=SAMPLE_LEN)
print(f'3 dof: {random_m=}')
random_c = np.random.randint(31, 40, size=SAMPLE_LEN)
print(f'3 dof: {random_c=}')

for i in range(SAMPLE_LEN):
    plt.plot(xx,f(random_m[i],xx,random_c[i]),color='yellow', label='Three degrees of freedom (m,x,c)' if i == 0 else "")


plt.ylim(first_output-10, 300)
plt.xlim(-5,10)
plt.legend()
plt.title('f(x) = mx + c')
plt.show()

# Note how the 1 degree of freedom (x) only goes in one y intercept and straight line
# The other lines with more degrees of freedom can have different trajectories and cross x=0 at different y-axis values
```
