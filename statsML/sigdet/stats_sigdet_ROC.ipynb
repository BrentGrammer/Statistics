```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
## first, re-create the dp and rb matrices from previous lectures

x  = np.arange(.01,1,.01)
# d-prime and response bias matrices
dp = np.tile(stats.norm.ppf(x),(99,1)).T - np.tile(stats.norm.ppf(x),(99,1))
rb = -( np.tile(stats.norm.ppf(x),(99,1)).T + np.tile(stats.norm.ppf(x),(99,1)) )/2
```

```python
## create the 2D bias spaces and plot their ROC curves

rb2plot = [.3, .5, .9, 1.5] # d-prime/response bias levels points which the curve maps along
# note that because the precision in the matrices produced above is high, we're not going to get values in the matrix that are exactly .3 or the values set here.
# The matricies are empirical and not derived from a formula

tol = .01 # tolerance for matching levels - we want points that are close to the rb2plot values (because of precision empirical matrix issue described)
colorz = 'rbmk'

# setup the figure
fig,ax = plt.subplots(1,2,figsize=(10,5))

# show the 2D spaces
ax[0].imshow(dp,extent=[x[0],x[-1],x[0],x[-1]],origin='lower')
ax[0].set_xlabel('False alarm rate')
ax[0].set_ylabel('Hit rate')
ax[0].set_title("d'")

ax[1].imshow(rb,extent=[x[0],x[-1],x[0],x[-1]],origin='lower')
ax[1].set_xlabel('False alarm rate')
ax[1].set_ylabel('Hit rate')
ax[1].set_title('Response bias')



### now draw the isocontours curves
for i in range(len(rb2plot)):
    
    # find d' points
    # find the points that are greater than the d-prime value within the tolerance range plus or minus (+/- .01)
    idx = np.where((dp>rb2plot[i]-tol) & (dp<rb2plot[i]+tol))
    ax[0].plot(x[idx[1]],x[idx[0]],'o-',color=colorz[i])
    
    # find bias points - same range as above +/- .01 of absolute values
    idx = np.where((rb>rb2plot[i]-tol) & (rb<rb2plot[i]+tol))
    ax[1].plot(x[idx[1]],x[idx[0]],'o-',color=colorz[i])


plt.show()

# note how the response bias curves are orthogonal/opposing in nature to the d-prime curves.
# also note as the d-prime value gets higher, the curve pulls towards the upper axes
  # The highest d-prime is the black left most curve (1.5)
# You need to take into account both the curve for d-prime and response bias to model every point in the signal detection theory space.
```
